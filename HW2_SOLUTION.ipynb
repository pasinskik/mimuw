```python
class SyntheticData[T](Dataset):
    """A small synthetic segmentation dataset.

    It is a sequence dataset of 5000 tuples (image, class, mask), where:
    - image: before transformation, an RGB PIL Image.
    - class: int 0..4, the label index.
    - mask:  numpy array of dtype=bool, shape (H, W), same size as image.
    """

    CLASSES: Final[tuple[str, ...]] = ("circle", "square", "triangle", "star", "diamond")

    def __init__(
        self,
        path: Path,
        split: Literal["train", "val", "test"],
        transform: Callable[[PIL.Image.Image], T],
    ) -> None:
        self.path = path
        self.transform = transform
        self.split = split

        self.class_to_idx = {class_name: i for i, class_name in enumerate(self.CLASSES)}

        images = sorted((self.path / "images" / self.split).glob("*.png"))
        masks = sorted((self.path / "masks" / self.split).glob("*.png"))
        labels = sorted((self.path / "labels" / self.split).glob("*.txt"))

        assert images, f"No images found in {self.path / 'images' / self.split}"
        assert len(images) == len(masks) == len(labels), (
            "Number of images, masks, and labels must be the same"
        )
        assert [p.stem for p in images] == [p.stem for p in masks] == [p.stem for p in labels], (
            "Image/mask/label filename mismatch."
        )
        self.image_names = [p.stem for p in images]

    def __len__(self) -> int:
        return len(self.image_names)

    def __getitem__(self, idx) -> tuple[T, int, np.ndarray]:
        name = self.image_names[idx]
        img_path = self.path / "images" / self.split / f"{name}.png"
        mask_path = self.path / "masks" / self.split / f"{name}.png"
        label_path = self.path / "labels" / self.split / f"{name}.txt"

        img = PIL.Image.open(img_path).convert("RGB")
        mask = np.array(PIL.Image.open(mask_path), dtype=bool)
        label = self.class_to_idx[label_path.read_text().strip()]

        img_transformed: T = self.transform(img)

        return img_transformed, label, mask


def show_image_row(
    image_dict: dict[str, PIL.Image.Image | np.ndarray | Tensor], size: float = 3.0
) -> None:
    n = len(image_dict)
    _, axs = plt.subplots(1, n, figsize=(size * n, size), constrained_layout=True, squeeze=True)
    if n == 1:
        axs = [axs]
    for ax, (title, img) in zip(axs, image_dict.items(), strict=True):
        ax.imshow(img)
        ax.set_title(title)
        ax.axis("off")
    plt.show()
```

```python
def train(
    model: nn.Module,
    train_loader: torch.utils.data.DataLoader,
    val_loader: torch.utils.data.DataLoader,
    test_loader: torch.utils.data.DataLoader,
    epochs: int = 5,
    **optimizer_kwargs: Any,
) -> None:
    print(f"ðŸš€ Training CNN for {epochs} epochs...")
    optimizer = optim.AdamW(model.parameters(), **optimizer_kwargs)

    for epoch in range(epochs):
        _train_epoch(model, train_loader, optimizer, desc=f"Epoch {epoch + 1}/{epochs} training  ")
        val_metrics = evaluate(model, val_loader, desc=f"Epoch {epoch + 1}/{epochs} validation")
        print(
            f"Epoch {epoch + 1}/{epochs} â€” "
            + f"val loss: {val_metrics['loss']:.3f}, val acc: {val_metrics['accuracy']:.1%}"
        )

    test_metrics = evaluate(model, test_loader, desc="Test Evaluation")
    print(
        "âœ… Model training complete: "
        + f"Test loss: {test_metrics['loss']:.3f}, test acc: {test_metrics['accuracy']:.1%}"
    )


def _train_epoch(
    model: nn.Module,
    dataloader: torch.utils.data.DataLoader,
    optimizer: optim.Optimizer,
    desc: str,
) -> dict[str, float]:
    model.train()
    device = next(model.parameters()).device
    total_loss = 0
    total_correct = 0
    total_samples = 0
    progress_bar = tqdm(dataloader, desc=desc)
    for imgs, labels, _ in progress_bar:
        imgs, labels = imgs.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(imgs)
        loss = nn.CrossEntropyLoss()(outputs, labels)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()

        _, predicted = torch.max(outputs.data, dim=1)
        total_samples += labels.shape[0]
        total_correct += (predicted == labels).sum().item()

        progress_bar.set_postfix(
            train_loss=f"{total_loss / (total_samples / labels.shape[0]):.3f}",
            train_acc=f"{total_correct / total_samples:.1%}",
        )

    return {"loss": total_loss / len(dataloader), "accuracy": total_correct / total_samples}


def evaluate(
    model: nn.Module, dataloader: torch.utils.data.DataLoader, desc: str
) -> dict[str, float]:
    model.eval()
    device = next(model.parameters()).device
    total_loss = 0
    total_correct = 0
    total_samples = 0
    with torch.no_grad():
        for imgs, labels, _mask in tqdm(dataloader, desc=desc):
            imgs, labels = imgs.to(device), labels.to(device)
            outputs = model(imgs)
            loss = nn.CrossEntropyLoss()(outputs, labels)
            total_loss += loss.item()
            _, predicted = torch.max(outputs.data, dim=1)
            total_samples += labels.shape[0]
            total_correct += (predicted == labels).sum().item()

    return {"loss": total_loss / len(dataloader), "accuracy": total_correct / total_samples}


class DataloaderArgs(TypedDict, total=False):
    batch_size: int
    shuffle: bool
    num_workers: int
    pin_memory: bool
```

## GradCAM implementation (add your code)

```python
class GradCAM:
    """
    Pure PyTorch implementation of Grad-CAM.

    Usage:
        grad_cam = GradCAM(model=model, target_layers=[layer1, layer2])
        grayscale_cam = grad_cam(input_tensor, targets=[class_id])

    High-level idea:
        Grad-CAM computes a class-specific localization map by:
        1) collecting feature maps from a convolutional layer,
        2) weighting them by gradients of the target class score,
        3) summing them and applying ReLU.
    """

    def __init__(self, model: nn.Module, target_layers: Iterable[nn.Module]) -> None:
        # Store model and target layers
        self.model = model
        self.target_layers = list(target_layers)

        # Will hold forward activations and backward gradients
        self._activations: list[Tensor] = []
        self._gradients: list[Tensor] = []

        # Keep track of hooks to remove them later
        self._hooks: list[Any] = []

        # Register hooks on target layers
        for layer in self.target_layers:
            self._hooks.append(layer.register_forward_hook(self._forward_hook))
            self._hooks.append(layer.register_full_backward_hook(self._backward_hook))

    def _forward_hook(self, module: nn.Module, inputs: tuple[Tensor, ...], output: Tensor) -> None:
        """
        This hook runs during the forward pass.

        - 'output' is the feature map tensor produced by the layer.
        - We save a clone of it to ensure its content is preserved.
        """
        self._activations.append(output.clone())

    def _backward_hook(self, module: nn.Module, grad_input: tuple[Tensor, ...], grad_output: tuple[Tensor, ...]) -> None:
        """
        This hook runs during the backward pass.

        - 'grad_output' contains the gradients with respect to the module's output.
        - We save a clone of the first gradient tensor.
        """
        # grad_output is typically a tuple, we are interested in the gradient w.r.t. the output feature map
        self._gradients.append(grad_output[0].clone())

    def _clear(self) -> None:
        """Clear stored activations and gradients before a new forward pass."""
        self._activations.clear()
        self._gradients.clear()

    def close(self) -> None:
        """Remove all hooks to avoid memory leaks."""
        for h in self._hooks:
            h.remove()
        self._hooks.clear()

    @torch.no_grad()
    def _normalize_per_sample(self, cam: Tensor) -> Tensor:
        """
        Normalize Grad-CAM maps independently for each image in the batch.

        Input:  cam shape (B, H, W)
        Output: cam shape (B, H, W), values in [0, 1]
        """
        B = cam.shape[0]

        # Flatten spatial dimensions to compute min/max per sample
        cam_flat = cam.view(B, -1)

        cam_min = cam_flat.min(dim=1).values.view(B, 1, 1)
        cam_max = cam_flat.max(dim=1).values.view(B, 1, 1)

        return (cam - cam_min) / (cam_max - cam_min + 1e-8)

    def __call__(self, input_tensor: Tensor, targets: Iterable[int] | None = None) -> np.ndarray:
        """
        Compute Grad-CAM heatmaps.

        Args:
        -----
        input_tensor: (B, 3, H, W) normalized images
        targets: optional iterable of target class indices (length B)

        Returns:
        --------
        numpy array of shape (B, H, W), values in [0, 1]
        """
        self.model.eval()
        self._clear()

        device = next(self.model.parameters()).device
        x = input_tensor.to(device)

        # Enable gradient tracking on inputs (needed for backpropagation)
        # Even though model is in eval mode, we need gradients for Grad-CAM's backward pass.
        with torch.enable_grad():
            x = x.requires_grad_(True)

            # Forward pass
            logits = self.model(x)  # shape: (B, num_classes)

            B, _, H, W = x.shape

            # If targets not provided, use predicted classes
            if targets is None:
                target_idxs = logits.argmax(dim=1)
            else:
                target_idxs = torch.tensor(list(targets), device=device, dtype=torch.long)
                assert target_idxs.shape == (B,), f"targets should have shape ({B},), got {target_idxs.shape}"

            # Backpropagate to get gradients of target class scores
            self.model.zero_grad(set_to_none=True)
            score = logits[torch.arange(B, device=device), target_idxs].sum()
            score.backward(retain_graph=False)

        # IMPORTANT: gradients were collected in backward order; activations in forward order.
        # Because we append gradients when they arrive, and multiple layers are used, we align by reversing gradients.
        grads = list(reversed(self._gradients))
        acts = self._activations

        assert len(acts) == len(grads) == len(self.target_layers), (
            f"Expected {len(self.target_layers)} layers worth of activations/gradients, "
            f"got activations={len(acts)}, gradients={len(grads)}"
        )

        cams_per_layer: list[Tensor] = []

        for A, G in zip(acts, grads, strict=True):
            # A: activations, shape (B, K, h, w)
            # G: gradients,   shape (B, K, h, w)

            # Compute weights: global average pooling of gradients
            weights = G.mean(dim=(2, 3), keepdim=True)  # (B, K, 1, 1)

            # Compute weighted combination of feature maps
            cam = (weights * A).sum(dim=1)              # (B, h, w)

            # Apply ReLU to keep only positive influences
            cam = F.relu(cam)

            # Upsample CAM to input image size
            cam = F.interpolate(cam.unsqueeze(1), size=(H, W), mode="bilinear", align_corners=False)
            cam = cam.squeeze(1)  # (B, H, W)
            cams_per_layer.append(cam)

        # Average CAMs from all target layers
        cam = torch.stack(cams_per_layer, dim=0).mean(dim=0)  # (B, H, W)

        # Normalize per image
        cam = self._normalize_per_sample(cam)

        return cam.detach().cpu().numpy()
```

```python
def heatmap_to_rgb_image(
    heatmap: np.ndarray, min: float | None = None, max: float | None = None
) -> PIL.Image.Image:
    """
    Converts a single-channel heatmap to an RGB pillow image using a colormap.

    Args:
    - heatmap: shape (H, W), will be normalized by mapping min..max to 0..1.
    - min: minimum value for normalization, defaults to heatmap.min().
    - max: maximum value for normalization, defaults to heatmap.max()
    """
    heatmap = heatmap.astype(np.float32)
    if min is None:
        min = heatmap.min()
    if max is None:
        max = heatmap.max()
    heatmap = (heatmap - min) / (max - min + 1e-8)
    heatmap_uint8 = (np.clip(heatmap, 0.0, 1.0) * 255).astype(np.uint8)
    heatmap_color = cv2.applyColorMap(heatmap_uint8, cv2.COLORMAP_JET)
    heatmap_color = cv2.cvtColor(heatmap_color, cv2.COLOR_BGR2RGB)
    return PIL.Image.fromarray(heatmap_color)
```

```python
class BasicSamPipeline:
    def __call__(self, images: Tensor) -> Tensor:
        """
        Input: normalized images, shape (B, C=3, H, W).
        Output: masks tensor of shape (B, H, W), dtype=bool.
        """
        B, C, H, W = images.shape
        # The basic pipeline always uses a single center point for each image.
        point_coords = np.array([[(W // 2, H // 2)] for _ in range(B)])

        # The basic pipeline always uses a single foreground point, no background points.
        point_labels = np.array([[1] for _ in range(B)], dtype=np.int64)

        return self.segment(images, point_coords, point_labels)

    def segment(
        self, images: Tensor, point_coords: np.ndarray, point_labels: np.ndarray
    ) -> Tensor:
        """
        Args:
        - images: normalized images, shape (B, C=3, H, W).
        - point_coords: point coordinates within each image, shape (B, num_points, 2), format (x,y).
            Note the format is not (h,w)=(y,x), but (x,y)!
        - point_labels: point labels, shape (B, num_points), dtype int64.
            Label 1 is foreground (should be in mask), 0 is background (shouldn't be in mask).

        Returns: segmentation masks, shape (B, H, W), dtype=bool.
        """
        B, C, H, W = images.shape
        assert C == 3, f"Expected images.shape=(B, C=3, H, W), got: {images.shape}"
        num_points = point_coords.shape[1]
        assert point_coords.shape == (B, num_points, 2), f"Expected point_coords.shape=({B=}, num_points, 2), got: {point_coords.shape}"
        assert point_labels.shape == (B, num_points), f"Expected point_labels.shape=({B=}, num_points), got: {point_labels.shape}"

        results = list[Tensor]()
        for image, pt_coords, pt_labels in zip(images, point_coords, point_labels, strict=True):
            sam_predictor.set_image(np.array(inverse_transform(image)))
            masks, scores, _logits = sam_predictor.predict(
                point_coords=pt_coords, point_labels=pt_labels, multimask_output=True
            )
            best_mask = masks[np.argmax(scores)]
            results.append(torch.tensor(best_mask, dtype=torch.bool))
        return torch.stack(results)
```

```python
def cam_to_points(
    cam: np.ndarray,
    num_fg: int = 1,
    num_bg: int = 0,
) -> tuple[np.ndarray, np.ndarray]:
    """
    Convert a Grad-CAM heatmap into SAM point prompts.

    Args:
        cam:
            Grad-CAM maps, shape (B, H, W), values in [0, 1].
            Higher value = more evidence for the target class at that location.
        num_fg:
            How many foreground points to generate per image.
            Foreground points (label=1) should lie *inside* the object.
        num_bg:
            How many background points to generate per image.
            Background points (label=0) should lie *outside* the object.

    Returns:
        point_coords:
            Array of shape (B, P, 2) where P = num_fg + num_bg.
            SAM expects (x, y) coordinates (IMPORTANT: not (y, x)).
        point_labels:
            Array of shape (B, P) with labels:
              1 = foreground (should be included in mask)
              0 = background (should be excluded from mask)
    """
    B, H, W = cam.shape
    P = num_fg + num_bg

    # Preallocate arrays for SAM prompts
    point_coords = np.zeros((B, P, 2), dtype=np.int64)
    point_labels = np.zeros((B, P), dtype=np.int64)

    # Process each image independently
    for b in range(B):
        heat = cam[b] # shape: (H, W)

        # Flatten heatmap so we can use fast "top-k" selection on 1D array
        flat = heat.reshape(-1)  # length H*W

        ### Foreground points (label = 1) ###
        # We want points where Grad-CAM is highest: "most evidence for the class".
        # np.argpartition is efficient: finds top-k without fully sorting the whole array.
        # Using -flat because we want largest values.
        top_idx = np.argpartition(-flat, kth=min(num_fg, flat.size - 1))[:num_fg]

        # Sort the selected indices by score descending for determinism/stability
        top_idx = top_idx[np.argsort(-flat[top_idx])]

        for i, idx in enumerate(top_idx):
            # Convert 1D index -> (y, x) in the heatmap grid
            y, x = divmod(int(idx), W)

            # SAM uses (x, y) coordinate order
            point_coords[b, i] = (x, y)
            point_labels[b, i] = 1  # foreground

        ### Background points (label = 0) ###
        # Background points can help SAM avoid "spilling" into large irrelevant areas.
        # We choose lowest-activation pixels: least evidence for the class.
        if num_bg > 0:
            low_idx = np.argpartition(flat, kth=min(num_bg, flat.size - 1))[:num_bg]
            low_idx = low_idx[np.argsort(flat[low_idx])]  # ascending (lowest first)

            for j, idx in enumerate(low_idx):
                y, x = divmod(int(idx), W)
                point_coords[b, num_fg + j] = (x, y)
                point_labels[b, num_fg + j] = 0  # background

    return point_coords, point_labels


class GradCamSamPipeline(BasicSamPipeline):
    """
    Pipeline: (images) -> Grad-CAM heatmap -> point prompts -> SAM mask

    We reuse BasicSamPipeline.segment(...) exactly as provided,
    but instead of using the image center as a prompt, we compute prompts from Grad-CAM.
    """
    def __init__(self, grad_cam: GradCAM, num_fg: int = 1, num_bg: int = 0):
        self.grad_cam = grad_cam
        self.num_fg = num_fg
        self.num_bg = num_bg

    def __call__(self, images: Tensor, targets: Tensor | None = None) -> Tensor:
        """
        Args:
            images:
                Normalized images (B, 3, H, W), using ImageNet mean/std (same as classifier input).
            targets:
                Optional tensor of shape (B,) with class labels.
                If None, Grad-CAM uses the classifier's predicted class per image.

        Returns:
            SAM-predicted masks of shape (B, H, W), dtype=bool.
        """
        # 1) Compute Grad-CAM heatmaps (B, H, W) in [0, 1]
        if targets is None:
            cams = self.grad_cam(images)
        else:
            cams = self.grad_cam(images, targets=targets.tolist())

        # 2) Convert heatmaps to SAM point prompts
        point_coords, point_labels = cam_to_points(
            cams, num_fg=self.num_fg, num_bg=self.num_bg
        )

        # 3) Run SAM with those prompts (segment() comes from BasicSamPipeline)
        return self.segment(images, point_coords, point_labels)
```

```python
def center_of_mass(mask: np.ndarray) -> tuple[float, float]:
    """Return (x,y) float center of mass of boolean mask. If mask is empty, fall back to image center."""
    ys, xs = np.where(mask)
    H, W = mask.shape
    if ys.size == 0:
        return (W / 2.0, H / 2.0)
    return (float(xs.mean()), float(ys.mean()))

@torch.no_grad()
def evaluate_prompts(pipeline_name: str, pipeline: Any, loader: DataLoader, desc: str, grad_cam: GradCAM | None = None) -> dict[str, float]:
    """
    Evaluate point prompts produced by a pipeline (basic or gradcam) using three metrics:
      - hit_rate: fraction of images where at least one foreground point falls inside GT mask,
      - mean_distance: mean (px) of distance from GT center-of-mass to the nearest FG point,
      - iou: mean Intersection-over-Union between pipeline predicted mask and GT mask.

    The function constructs prompts (either center or from Grad-CAM), calls pipeline.segment(...)
    to obtain predicted masks and computes per-image metrics that are averaged over the dataset.
    """
    hits: list[int] = []
    distances: list[float] = []
    ious: list[float] = []

    for images, labels, gt_masks in tqdm(loader, desc=desc):
        B, C, H, W = images.shape

        if pipeline_name == 'basic':
            point_coords = np.array([[(W // 2, H // 2)] for _ in range(B)])
            point_labels = np.array([[1] for _ in range(B)], dtype=np.int64)
        elif pipeline_name == 'gradcam':
            assert grad_cam is not None, 'grad_cam must be provided for gradcam pipeline evaluation'
            cams = grad_cam(images.to(device))  # (B, H, W) values in [0,1]
            point_coords, point_labels = cam_to_points(cams, num_fg=pipeline.num_fg, num_bg=pipeline.num_bg)
        else:
            raise ValueError('Unknown pipeline_name: ' + str(pipeline_name))

        images_cpu = images.cpu()
        pred_masks = pipeline.segment(images_cpu, point_coords, point_labels).cpu()  # (B,H,W), bool

        # For hit rate / distance / IoU: check each image in the batch
        for b in range(B):
            gt = gt_masks[b]  # numpy bool (H, W)
            gt_t = torch.tensor(gt, dtype=torch.bool)

            fg_idx = np.where(point_labels[b] == 1)[0]
            hit = 0
            min_dist = float('nan')
            com_x, com_y = center_of_mass(gt)
            if fg_idx.size > 0:
                dists = []
                for idx in fg_idx:
                    x, y = int(point_coords[b, idx, 0]), int(point_coords[b, idx, 1])
                    x = int(np.clip(x, 0, W - 1))
                    y = int(np.clip(y, 0, H - 1))
                    if gt[y, x]:
                        hit = 1
                    d = float(np.sqrt((x - com_x) ** 2 + (y - com_y) ** 2))
                    dists.append(d)
                min_dist = float(np.min(dists))
            else:
                hit = 0
                min_dist = float(np.sqrt(H ** 2 + W ** 2))

            # IoU between predicted mask and GT
            pred = pred_masks[b].bool()
            inter = float(((pred) & (gt_t)).sum().item())
            union = float(((pred) | (gt_t)).sum().item())
            iou = inter / (union + 1e-8)

            hits.append(hit)
            distances.append(min_dist)
            ious.append(iou)

    return {"hit_rate": float(np.mean(hits)), "mean_distance": float(np.mean(distances)), "iou": float(np.mean(ious))}

basic_pipeline = BasicSamPipeline()
grad_cam = GradCAM(model=model, target_layers=target_layers)
# First approach: 1 FG point only
gradcam_sam = GradCamSamPipeline(grad_cam=grad_cam, num_fg=1, num_bg=0)
# Second approach: 1 FG point + 1 BG point
gradcam_bg_sam = GradCamSamPipeline(grad_cam=grad_cam, num_fg=1, num_bg=1)

basic_metrics = evaluate_prompts('basic', basic_pipeline, sam_eval_loader, desc="Basic SAM (center point) prompts")
gradcam_metrics = evaluate_prompts('gradcam', gradcam_sam, sam_eval_loader, desc="Grad-CAM -> SAM (1 FG) prompts", grad_cam=grad_cam)
gradcam_bg_metrics = evaluate_prompts('gradcam', gradcam_bg_sam, sam_eval_loader, desc="Grad-CAM -> SAM (1 FG + 1 BG) prompts", grad_cam=grad_cam)

print("Basic SAM prompts:", basic_metrics)
print("Grad-CAM -> SAM (1 FG) prompts:", gradcam_metrics)
print("Grad-CAM -> SAM (1 FG + 1 BG) prompts:", gradcam_bg_metrics)
```

```python
def _draw_points_on_pil(img: PIL.Image.Image, point_coords: np.ndarray, point_labels: np.ndarray, radius: int = 5) -> PIL.Image.Image:
    """Draw point prompts on a copy of PIL image. point_coords shape=(P,2) in (x,y) order."""
    img = img.copy()
    draw = PIL.ImageDraw.Draw(img)
    P = point_coords.shape[0]
    for p in range(P):
        x, y = int(point_coords[p, 0]), int(point_coords[p, 1])
        label = int(point_labels[p])
        color = (0, 255, 0) if label == 1 else (255, 0, 0)
        draw.ellipse((x - radius, y - radius, x + radius, y + radius), outline=color, width=2)
    return img

def visualize_pipeline_comparison(indices=(3, 10, 42)):
    images, labels, gt_masks = next(iter(test_loader))
    images, labels, gt_masks = images[list(indices)], labels[list(indices)], gt_masks[list(indices)]

    grad_cam_local = GradCAM(model=model, target_layers=target_layers)
    cams = grad_cam_local(images.to(device), targets=labels.tolist())

    basic = BasicSamPipeline()
    gc_pipeline = GradCamSamPipeline(grad_cam_local, num_fg=1, num_bg=0)
    gc_bg_pipeline = GradCamSamPipeline(grad_cam_local, num_fg=1, num_bg=1)

    basic_pred = basic(images).cpu()
    gc_pred = gc_pipeline(images, targets=labels).cpu()
    gc_bg_pred = gc_bg_pipeline(images, targets=labels).cpu()

    B, C, H, W = images.shape

    # Extract points for all approaches
    basic_point_coords = np.array([[(W // 2, H // 2)] for _ in range(B)])
    basic_point_labels = np.ones((B, 1), dtype=np.int64)

    gc_point_coords, gc_point_labels = cam_to_points(cams, num_fg=1, num_bg=0)
    gc_bg_point_coords, gc_bg_point_labels = cam_to_points(cams, num_fg=1, num_bg=1)

    for i in range(B):
        heatmap_img = heatmap_to_rgb_image(cams[i], 0, 1)
        inp = inverse_transform(images[i])

        # Draw points (Green = FG, Red = BG)
        inp_basic = _draw_points_on_pil(inp, basic_point_coords[i], basic_point_labels[i])
        inp_gc = _draw_points_on_pil(inp, gc_point_coords[i], gc_point_labels[i])
        inp_gc_bg = _draw_points_on_pil(inp, gc_bg_point_coords[i], gc_bg_point_labels[i])

        show_image_row({
            "Input": inp,
            "Grad-CAM heatmap": heatmap_img,
            "GT mask": gt_masks[i],
            "Basic Point": inp_basic,
            "GC Point (1FG)": inp_gc,
            "GC Points (1FG+1BG)": inp_gc_bg,
            "Basic SAM": basic_pred[i],
            "GC SAM (1FG)": gc_pred[i],
            "GC SAM (1FG+1BG)": gc_bg_pred[i],
        }, size=2.5) # Smaller size to fit more columns

visualize_pipeline_comparison()
```
